# Creating, Reading and Writing

### DataFrames

A DataFrame is a table.

```python
pd.DataFrame({'Column A': ['entry a-0', 'entry a-1'], 
              'Column B': ['entry b-0', 'entry b-1']}, 
             index = ['index 0', 'index 1'])
```

We are using the `pd.DataFrame()` constructor to generate these DataFrame objects, where the keys are the column names, and the values are a list of entries.

The list of row labels used in a DataFrame is known as an **Index** - the default values are ascending numbers starting from 0 - we can assign values to it by using an index parameter in our constructor:

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
```

### Series

A Series is a sequence of data values, **a single column** of a DataFrame.


```python
s1 = pd.Series([1, 2, 3, 4, 5])
print(f'Series without index and name:\n{s1}')

# with index name:
s2 = pd.Series([30, 35, 40], index=['2015 Sales', '2016 Sales', '2017 Sales'], name='Product A')

print(f'\nSeries with index and name:\n{s2}')
```

# Indexing, Selecting and Assigning

This segment helps learn how to select specific values out of the data. 

### Naive accessors

The indexing operator and attribute selection work like they do in the Python ecosystem - Pandas allows to access property of data by **accessing the object attribute** either by dot notation or by bracket notation

```python
print(wine_reviews.country)  # same as reviews['country']

print(f'\ncountry[0]: {wine_reviews.country[0]}')
```

### Indexing

Pandas also has it's own accessor operators, `loc` and `iloc`, which are used for **more advance operations**.

Both accessor operators are **row-first, column-second**, as opposed to native Python.

### Index-based Selection - Numeric Position

selecting data based on its numerical position with `iloc`.

```python
wine_reviews.iloc[0]  # get first row (data for Italy)
```

To get column data, use slicing (similar to Python accessors):

```python
wine_reviews.iloc[:, 0]  # all country data (all data of index 'country')
```

It's also possible to pass a list `reviews.iloc[[0, 1, 2], 0]`, or use negative numbers `reviews.iloc[-5:]`

```python
print(reviews.iloc[[0, 1, 2], 0])

reviews.iloc[-5:]
```

### Label-based Selection - Index Value

selecting data based on its index value with `loc`.

```python
wine_reviews.loc[0] # get first row (data for Italy)
```

```python
print(wine_reviews.loc[0, 'country'])
print(wine_reviews.iloc[0, 0])
```

```python
wine_reviews.loc[:, ['country', 'taster_name']]  # get all countries and taster names
```

### Using `loc` vs. `iloc`

`iloc`
1. conceptually simpler, ignores the dataset's indices and treats it as a big matrix
2. Uses Python stdlib indexing scheme - first element of a range is included and the last excluded

`loc`
1. uses the information in the indices, more meaningful indices
2. Uses inclusive indices, first and last elements of range are included  (since values of range are chosen by *value*)

## Conditional Selection

We'd often need to ask questions based on conditions. Suppose we're interested in better-than-average wines produced in Italy:

```python
# check if country is Italy
wine_reviews.loc[reviews.country == 'Italy']
```

```python
# use ampersand `&` or pipe `|` to bring two questions together
wine_reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 99)]
```

Pandas also has built-in conditional selectors such as `isin`, `isnull`, `notnulll`

```python
reviews.loc[reviews.country.isin(['Italy', 'France'])]  # output all reviews where country is in the given list.
```

```python
reviews.loc[reviews.price.notnull()]  # output all reviews with price not null.
```

## Maps

A term borrowed from mathematics, maps one set of values to another set of values.

1. `map()` - a slightly simpler mapper, maps Series values
2. `apply()` - the equivalent method, transforms a whole DataSeries by calling a custom method on each row
3. Operators - faster yet less flexible

Both methods are rerturning **new objects** and do not affect the existing Series or DataFrame.

```python
review_points_mean = wine_reviews.points.mean()
wine_reviews.points
```

```python
wine_reviews.points.map(lambda p: p - review_points_mean)
```

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews.apply(remean_points, axis='columns')
```

```python
reviews.taster_name + ", from: " + reviews.country
```

```python
tropical_count = reviews.description.map(lambda desc: 'tropical' in desc).sum()
fruity_count = reviews.description.map(lambda desc: 'fruity' in desc).sum()

descriptor_counts = pd.Series([tropical_count, fruity_count], index=['tropical', 'fruity'])
print(descriptor_counts)
```

# Grouping and Sorting

* Grouping by single-index or multi-index (a common "gotcha" in pandas)
    * Multi-indices have a **tiered structure**, and require two levels of labels to retrieve a value [see pandas docs - Advanced](https://pandas.pydata.org/pandas-docs/stable/advanced.html)
    * The order is dependant on the value of the index (and not the data)
* Aggregating data using `agg()`
* Reset the index using `reset_index()`

```python
wine_reviews.groupby(['country']).price.agg([len, min, max])
```
